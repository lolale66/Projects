```python

from datetime import datetime
import math

# These are random values just for illustration
STARTING_CAR_NUMBER = 10
HOURLY_RATE = 10
DAILY_RATE = 100
WEEKLY_RATE = 500


class RentCar:
    def __init__(self, available_cars):
        self.available_cars = available_cars

    def display_available_cars(self):
        print(f'There are {self.available_cars} available cars')

    def rent_car_hourly(self, requested_cars):
        if requested_cars < 0:
            print(f'Number of requested cars should be positive')
        elif requested_cars > self.available_cars:
            print(f'Not enough cars available.')
        else:
            self.rental_time = datetime.now()
            self.available_cars -= requested_cars

    def rent_car_daily(self, requested_cars):
        if requested_cars < 0:
            print(f'Number of requested cars should be positive')
        elif requested_cars > self.available_cars:
            print(f'Not enough cars available.')
        else:
            self.rental_time = datetime.now()
            self.available_cars -= requested_cars

    def rent_car_weekly(self, requested_cars):
        if requested_cars < 0:
            print(f'Number of requested cars should be positive')
        elif requested_cars > self.available_cars:
            print(f'Not enough cars available.')
        else:
            self.rental_time = datetime.now()
            self.available_cars -= requested_cars

    def return_car(self, rental_time, rental_mode, number_of_cars):
        self.available_cars += number_of_cars
        # Calculate duration in seconds
        duration = datetime.now() - rental_time
        duration_in_s = duration.total_seconds()
        if rental_mode == 'hourly':
            hours = math.ceil(duration_in_s / (60 * 60))
            print(f'Your bill is ${hours * HOURLY_RATE * number_of_cars}')
        elif rental_mode == 'daily':
            days = math.ceil(duration_in_s / (60 * 60 * 24))
            print(f'Your bill is ${days * DAILY_RATE *number_of_cars}')
        elif rental_mode == 'weekly':
            weeks = math.ceil(duration_in_s / (60 * 60 * 24 * 7))
            print(f'Your bill is ${weeks * WEEKLY_RATE * number_of_cars}')


class Customer:
    def __init__(self, name):
        self.name = name

    def request_car(self, rental_time, rental_mode, requested_cars, car_rental):
        if rental_mode == 'hourly':
            car_rental.rent_car_hourly(requested_cars)
        elif rental_mode == 'daily':
            car_rental.rent_car_daily(requested_cars)
        elif rental_mode == 'weekly':
            car_rental.rent_car_weekly(requested_cars)

    def return_car(self, rental_time, rental_mode, number_of_cars, car_rental):
        car_rental.return_car(rental_time, rental_mode, number_of_cars)


def main():
    available_cars = STARTING_CAR_NUMBER
    car_rental = RentCar(available_cars)
    customer = Customer("John")

    while True:
        print("1. Display available cars")
        print("2. Rent a car")
        print("3. Return a car")
        print("4. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            car_rental.display_available_cars()
        elif choice == "2":
            rental_time = datetime.now()
            rental_mode = input("Enter rental mode (hourly, daily, weekly): ")
            num_cars = int(input("Enter the number of cars to rent: "))
            customer.request_car(rental_time, rental_mode, num_cars, car_rental)
        elif choice == "3":
            rental_time = datetime.now()
            rental_mode = input("Enter rental mode (hourly, daily, weekly): ")
            num_cars = int(input("Enter the number of cars to return: "))
            customer.return_car(rental_time, rental_mode, num_cars, car_rental)
        elif choice == "4":
            break
        else:
            print("Invalid choice. Please try again.")


if __name__ == "__main__":
    main()
```
